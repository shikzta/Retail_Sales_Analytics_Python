```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../01_Data/01_Raw/retail_sales.csv")
df.head()               
```

```python
df.shape
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['order_date']=pd.to_datetime(df['order_date'])
df.dtypes
```

```python
df.head()
```

```python
df['year']=df['order_date'].dt.year
```

```python
df.head()
```

```python
df['month']=df['order_date'].dt.month_name()
```

```python
df.head()
```

```python
df['day']=df['order_date'].dt.day
```

```python
df.head()
```

```python
df['weekday']=df['order_date'].dt.day_name()
```

```python
df.head()
```

```python
df['revenue']=df['quantity']*df['unit_price']
```

```python
df.head()
```

```python
df['Total_Revenue']=df['revenue'].sum()
```

```python
df.head()
```

```python
df.head()
```

```python
revenue_by_category=df.groupby('category')['revenue'].sum().sort_values(ascending=False)
```

```python
revenue_by_category
```

```python
df.columns
```

```python
df.drop(columns=['revenue_by_category'], inplace=True)
```

```python
df.head()
```

```python
revenue_by_category
```

```python
plt.bar(revenue_by_category.index,revenue_by_category.values)
plt.title('Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Revenue by Product Category')
bars=plt.bar(revenue_by_category.index, revenue_by_category.values)
for bar in bars:
    height=bar.get_height()
    x=bar.get_x()+bar.get_width()/2
    plt.text(x,height+2,x+height,ha='center')
plt.savefig('../04_Outputs/01_Charts/revenue_by_category.png')
plt.show()
```

```python
revenue_by_month=df.groupby('month')['revenue'].sum()
```

```python
revenue_by_month
```

```python
plt.bar(revenue_by_month.index,revenue_by_month.values)
plt.title('Revenue By Month')
plt.xlabel('Month')
plt.ylabel('Revenue')
bars=plt.bar(revenue_by_month.index, revenue_by_month.values)
for bar in bars:
    height=bar.get_height()
    x=bar.get_x()+bar.get_width()/2
    plt.text(x,height+2,x+height,ha='center')
plt.savefig('../04_Outputs/01_Charts/revenue_by_month.png')
```

```python
revenue_by_country=df.groupby('country')['revenue'].sum().sort_values(ascending=False)
```

```python
revenue_by_country
```

```python
plt.bar(revenue_by_country.index, revenue_by_country.values)
plt.title('Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Revenue')
bars=plt.bar(revenue_by_country.index, revenue_by_country.values)
for bar in bars:
    height=bar.get_height()
    x=bar.get_x()+bar.get_width()/2
    plt.text(x,height+2,x+height,ha='center')
plt.savefig('../04_Outputs/01_Charts/revenue_by_country.png')
```

```python
revenue_by_weekday=df.groupby('weekday')['revenue'].sum()
plt.xticks(rotation=45, ha='right')
revenue_by_weekday=revenue_by_weekday.round(2)
order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
revenue_by_weekday=revenue_by_weekday.reindex(order)
plt.bar(revenue_by_weekday.index, revenue_by_weekday.values)
plt.title('Revenue by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Revenue')
bars=plt.bar(revenue_by_weekday.index, revenue_by_weekday.values)
for bar in bars:
    height=bar.get_height()
    x=bar.get_x()+bar.get_width()/2
    plt.text(
        x,
        height+2,
        round(height,2),
        ha='center',
        rotation=45
    )
plt.tight_layout()
plt.savefig('../04_Outputs/01_Charts/revenue_by_weekday.png')
```

```python
revenue_by_top5_product=df.groupby('product')['revenue'].sum().sort_values(ascending=False).head(5)
```

```python
revenue_by_top5_product=revenue_by_top5_product.round(2)
plt.bar(revenue_by_top5_product.index, revenue_by_top5_product.values)
plt.xticks(rotation=45, ha='right')
plt.title('Revenue by Top 5 Products')
plt.xlabel('Product')
plt.ylabel('Revenue')
bars=plt.bar(revenue_by_top5_product.index, revenue_by_top5_product.values)
for bar in bars:
    height=bar.get_height()
    x=bar.get_x()+bar.get_width()/2
    plt.text(
        x,
        height+2,
        round(height,2),
        ha='center',
        rotation=45
    )
plt.ylim(0, revenue_by_top5_product.max() * 1.18)
plt.tight_layout()
plt.savefig('../04_Outputs/01_Charts/revenue_by_top5_product.png')
```

```python
revenue_by_top5_customers=df.groupby('customer_id')['revenue'].sum().sort_values(ascending=False).head(5)
plt.bar(revenue_by_top5_customers.index, revenue_by_top5_customers.values)
plt.title('Revenue by Customer ID')
plt.xlabel('Customer ID')
plt.ylabel('Revenue')
bars=plt.bar(revenue_by_top5_customers.index,revenue_by_top5_customers .values)
for bar in bars:
    height=bar.get_height()
    x=bar.get_x()+bar.get_width()/2
    plt.text(
        x,
        height+2,
        round(height,2),
        ha='center',
        rotation=45
    )
plt.ylim(0, revenue_by_top5_product.max() * 1.66)
plt.tight_layout()
plt.savefig('../04_Outputs/01_Charts/revenue_by_top5_customers.png')
```

##### 

```python
customer_top6_purchase_freq=df['customer_id'].value_counts().head(6)
plt.bar(customer_top6_purchase_freq.index, customer_top6_purchase_freq.values)
plt.title('Purchase Frequency by Customer ID')
plt.xlabel('Customer ID')
plt.ylabel('Orders')
bars=plt.bar(customer_top6_purchase_freq.index,customer_top6_purchase_freq.values)
for bar in bars:
    height=bar.get_height()
    x=bar.get_x()+bar.get_width()/2
    plt.text(
        x,
        height+0.02,
        round(height,2),
        ha='center',
        rotation=45
    )
plt.ylim(0, customer_top6_purchase_freq.max() * 1.2)
plt.tight_layout()
plt.savefig('../04_Outputs/01_Charts/customer_top6_purchase_freq.png')
```
